## Investigation Overview

> This presentation aims at informing the audience what kinds of loans most people would normally opt for and recommend as it was discovered from the explorations made. Most people prefer the least risky kinds of loan as most of the recommendations made were for the least risky loans, which is no surprise. Also, medium term loans which have high APRs were preferred to long term loans.
 


## Dataset Overview

> There are 113,937 loans in the dataset with 81 features of each loan including; loan amount, borrower rate (or interest rate), current loan status, borrower income with most variables being numeric in nature. The aim of the exploration is to determine which of the features really affected the status of each loan. The features of interest in this exploration are Borrower's Annual Percentage Rate (APR), Borrower's interest rate for the loan and Term of loan.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df= pd.read_csv('prosperLoanData.csv')
```

## (Visualization 1)
### More medium term loans were preferred
##### >Below is the viualization of loan term versus borrower APR. As stated in the key insights above, low/medium term loans are preferred by borrowers compared to the long term loan because of its high borrower APR.

```python
color= sb.color_palette()[0]
#plotting borrower APR chart and loan term chart side by side for individual analysis
plt.figure(figsize=[20,5])
plt.subplot(1,2,1)
bins= np.arange(0,df['BorrowerAPR'].max()+0.02,0.02)
plt.hist(data=df, x='BorrowerAPR', bins=bins)
plt.grid()
plt.title('Annual Percentage Rate')
plt.xlabel('Borrower APR')
plt.ylabel('No of Borrowers')

#Plotting loan term chart
plt.subplot(1,2,2)
sb.countplot(data=df, x='Term', color=color)
plt.title('Loan Term')
plt.ylabel('No of Borrower');
```

### Which loan term is mostly taken by borrowers?

##### Most borrowers took loans with Annual Percentage Rates of 0.16 and 0.20 moreover, medium loan term is mostly opted for by borrowers as to be seen in the chart below

```python
#plotting the borrower APR versus the loan term. I multiplied the APRs by 50 to get better understanding of the chart
color= sb.color_palette()[0]
plt.figure(figsize=[20,10])
sb.boxplot(data=df, x='Term', y=df['BorrowerAPR']*50, color=color)
plt.title('Borrower APR versus Term of Loan')
plt.ylabel('Annual Percentage Rates');
```

##### Another point to note here, from the above plot is that the average number of borrowers that took loans with small term is higher than those with medium term loans. However, the number of borrowers with medium term loans is more than those with small or long term loans hence, corroborating the earlier stated fact that most borrowers opted for the medium term loan.

## (Visulization 2)

### Least Risky loans got very high recommendations



##### > The visulization below shows the relationship between recommendation and prosper score. Here we see that more people recommended the less risky kinds of loans which relates to why most borrowers opted for the low/medium term loans as explained in the previous visualization.


```python
plt.figure(figsize=[20,10])
sb.boxplot(data=df, x='Recommendations', y='ProsperScore', color=color)
plt.title('Recommendations and Prosper Score');
```

##### > Least risky loans had a high recommendation. This suggests that most people who have taken loans took loans with less risk and also recommended same 

## (Visualization 3)
### Medium term loans is high in all categories of loan status


##### >This visulization shows the relationship between Loan status and loan term. As seen in the exploratory part, more people actually went for the medium term loan and it is observed that medium term loans are high in all the categories of loan status.

```python
#plotting the loan status against term of loan
plt.figure(figsize=[15,10])
sb.countplot(data=df, x='LoanStatus', hue='Term', color=color)
plt.xticks(rotation=60)
plt.title('Loan Status versus Term')
plt.ylabel('No of Borrowers');
```

##### > Loans that are on-going seem to have more borrowers in the medium term loan but not as much as medim term borrowers who have completed their loans. However, in all of the loan status categories more people took the medium term loans.

## In Conclusion
> 
###### From the above explanations that more borrowers from all categories of loan status took the medium term loans. Moreover, the fact that most of the borrowers took the least risky loans with low annual percentage rates APRs and that medium term loans have an average annual percentage rate which is low invariably shows that the loan term, APR and Interest rate goes a long way to affect the status of loans.
